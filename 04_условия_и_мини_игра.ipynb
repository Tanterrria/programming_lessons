## Как это работает

В жизни:
- если идет дождь -> берем зонт,
- иначе -> идем без зонта.

В Python:
- `if` = если,
- `elif` = иначе если,
- `else` = иначе.

## Первый `if/else`

```python
age = int(input("Сколько тебе лет? "))

if age >= 18:
    print("Проходи, вот твой билет!")
else:
    print("Извини, этот фильм только для взрослых.")
```

## Несколько вариантов через `elif`

```python
age = int(input("Сколько тебе лет? "))

if age < 12:
    print("Детский билет со скидкой.")
elif age < 18:
    print("Подростковый билет, нужен взрослый.")
else:
    print("Взрослый билет.")
```

## Вложенные `if`

Иногда одной проверки мало. Тогда внутри `if` или `else` можно сделать еще одну проверку.

Это и есть вложенный `if`: проверка внутри проверки.

```python
age = int(input("Сколько тебе лет? "))

if age >= 18:
    print("Проходи, ты взрослый.")
else:
    print("Ты младше 18.")
    if age < 12:
        print("Тебе доступен детский билет.")
    else:
        print("Тебе доступен подростковый билет.")
```

## Операторы сравнения

- `>` больше
- `<` меньше
- `>=` больше или равно
- `<=` меньше или равно
- `==` равно
- `!=` не равно
