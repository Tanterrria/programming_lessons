## 1) Идея условий

В жизни мы постоянно делаем выбор:
- если идет дождь -> берем зонт,
- иначе -> идем без зонта.

В Python так же:
- `if` = если,
- `elif` = иначе если,
- `else` = иначе.

## 2) Первое условие

```python
age = int(input("Сколько тебе лет? "))

if age >= 18:
    print("Проходи, вот твой билет!")
else:
    print("Извини, этот фильм только для взрослых.")
```

Пояснение:
- `>=` означает "больше или равно".
- После условия ставим `:`.
- Внутри блока обязательно отступ (обычно 4 пробела).

## 3) Несколько условий подряд (`elif`)

```python
age = int(input("Сколько тебе лет? "))

if age < 12:
    print("Детский билет со скидкой.")
elif age < 18:
    print("Подростковый билет, нужен взрослый.")
else:
    print("Взрослый билет.")
```

## 4) Таблица операторов сравнения

- `>` больше
- `<` меньше
- `>=` больше или равно
- `<=` меньше или равно
- `==` равно
- `!=` не равно
