# 05. Списки в Python

В этом уроке разберём **список** (`list`): как хранить **несколько значений** под одним именем, как читать и менять элементы, как добавлять и убирать элементы из списка.

Списки пригодятся везде: инвентарь в игре, оценки за четверть, список имён героев, шаги маршрута — всё, где порядок важен и элементов может стать больше или меньше.

## 1) Что такое список

**Список** — это **упорядоченная** последовательность значений в **квадратных скобках** `[]`. Значения называют **элементами**.

- Порядок важен: первый элемент, второй, третий…
- В одном списке могут лежать **разные типы** (число, строка, даже другой список) — но чаще держат однотипные данные для ясности.
- Список **можно менять**: заменить элемент, добавить в конец, удалить — в отличие от строки, где отдельные символы «внутри» не перепишешь.

```python
# Примеры списков
scores = [10, 12, 9]
friends = ["Аня", "Вован", "Костя"]
mixed = [1, "два", True]

print("Очки:", scores)
print("Друзья:", friends)
print("Всё подряд:", mixed)
```

## 2) Как создать список

- **Сразу с элементами:** `nums = [1, 2, 3]`
- **Пустой список:** `items = []` или `items = list()`

Длину списка (сколько в нём элементов) узнают функцией **`len(...)`**.

```python
empty = []
print("Пустой список:", empty, "длина:", len(empty))

letters = ["a", "b", "c"]
print("Буквы:", letters, "длина:", len(letters))
```

## 3) Индекс: как взять один элемент

У каждого элемента есть **номер — индекс**. В Python счёт **с нуля**:

- первый элемент — индекс `0`,
- второй — `1`,
- последний можно взять как **`-1`**, предпоследний — `-2` и так далее.

Запись `имя[индекс]` читается как «взять элемент с этим номером».

```python
team = ["Линза", "Щит", "Стрела"]

print("Первый в списке:", team[0])
print("Второй:", team[1])
print("Последний:", team[-1])
```

## 4) Срез: кусок списка

Синтаксис **`список[от:до]`** — берёт подсписок **от** индекса `от` **до** (не включая) индекса `до`.

- Если опустить **левую** границу — считается с начала: `[:3]` — первые три элемента.
- Если опустить **правую** — до конца: `[1:]` — со второго элемента до конца.

Срез **не ломает** исходный список: получается новый (маленький) список для чтения или дальнейших действий.

```python
nums = [10, 20, 30, 40, 50]

print("Срез [1:4]:", nums[1:4])
print("С начала до индекса 3 (не включая 3):", nums[:3])
print("С индекса 2 до конца:", nums[2:])
```

## 5) Изменить элемент по индексу

Список **изменяемый**: можно записать новое значение на место старого:

`список[индекс] = новое_значение`

```python
hp = [100, 80, 50]
print("До лечения:", hp)

hp[1] = 100
print("После зелья второму герою:", hp)
```

## 6) Добавить и убрать элемент

- **`append(x)`** — добавить **`x` в конец** списка.
- **`insert(i, x)`** — вставить **`x` на позицию с индексом `i`** (остальные сдвигаются).
- **`pop()`** без аргумента — убрать и **вернуть** последний элемент.
- **`pop(i)`** — убрать элемент с индексом **`i`**.
- **`remove(x)`** — убрать **первое** вхождение значения **`x`** (по значению, не по индексу).

Если индекс выйдет за границы списка или элемента нет — Python выдаст ошибку: нужно следить за длиной и содержимым.

```python
bag = ["яблоко", "хлеб"]
print("Старт:", bag)

bag.append("сыр")
print("После append:", bag)

bag.insert(1, "масло")
print("После insert(1, ...):", bag)

last = bag.pop()
print("Достали с конца:", last, "-> осталось:", bag)

bag.remove("хлеб")
print("Убрали хлеб по значению:", bag)
```

## 7) Сложение и умножение списков

- **`+`** склеивает два списка **в один новый** (исходные не меняются, если не переприсвоить результат).
- **`*` с целым числом** — повторяет список заданное число раз: `[0] * 5` даёт пять нулей.

Это удобно для заготовок: например, `["?"] * 3` — три заглушки.

```python
a = [1, 2]
b = [3, 4]
print("Склейка:", a + b)

row = ["_"] * 5
print("Полоска из пяти ячеек:", row)
```

## 8) Проверка «есть ли в списке»

Оператор **`in`** возвращает `True` или `False`:

`значение in список`

Часто это используют вместе с условием `if` (ты это уже проходил в уроке про условия): например, проверить, есть ли предмет в инвентаре.

```python
inventory = ["меч", "щит", "зелье"]

print("меч в сумке?", "меч" in inventory)
print("лук в сумке?", "лук" in inventory)
```
