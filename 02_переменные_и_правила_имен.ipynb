## Переменная

**Переменная** — это имя, которому **присвоено** значение. По этому имени ты потом читаешь данные и можешь записать новое значение. Имя переменной всегда является идентификатором. В более крупных программах идентификаторы бывают ещё у функций, классов и т.д. — об этом позже.

В Python переменная ведёт себя как **ссылка на объект в памяти** (на конкретное значение), а тип данных Python обычно понимает по самому значению.

## 1) Что такое переменная

Переменная — это «подписанная коробка» для данных: у коробки есть **имя-идентификатор**, внутри лежит значение.

Например:
- `name` хранит имя,
- `age` хранит возраст.

```python
name = "Вован"
age = 13
print(name)
print(age)
```

## 2) Присваивание

- Переменная **создаётся** в момент первого присваивания (`=`).
- **Простое присваивание:** `a = 5`.
- **Одно значение нескольким именам:** `x = y = z = 0`.
- **Несколько имён за раз:** `a, b, c = 1, 2, 3` (порядок слева и справа должен совпадать).
- **Переприсваивание:** можно записать новое значение в то же имя; часто пишут короче: `n += 1` вместо `n = n + 1`.
- **Обмен двух значений** без третьей переменной: `a, b = b, a`.

```python
# Примеры присваивания
score = 10

# Одно значение нескольким именам
x = y = z = 0
print("x, y, z:", x, y, z)

# Несколько имён за раз
a, b, c = 1, 2, 3
print("a, b, c:", a, b, c)

# Переприсваивание
counter = 0
counter = counter + 1
print("counter:", counter)
counter += 1
print("counter:", counter)

# Обмен двух значений
first = 5
second = 10
print("first, second до обмена:", first, second)
first, second = second, first
print("first, second после обмена:", first, second)
```

Пример с именем и фамилией

Здесь главное увидеть **переприсваивание**: переменная может указывать на новое значение.

Сначала:
- `name = "Вован"`
- `last_name = "Вованов"`
- `name_2 = "Аня"`
- `last_name_2 = "Цветова"`

Потом меняем значение `last_name_2` (новая строка вместо старой).

```python
name = "Вован"
last_name = "Вованов"

name_2 = "Аня"
last_name_2 = "Цветова"

print("До:")
print(name, last_name)
print(name_2, last_name_2)

# Меняем значение переменной. Для девушек добавляем "a" к концу фамилии.
last_name_2 = last_name + 'a'

print("\nПосле:")
print(name_2, last_name_2)
```

## 3) Правила имён и ключевые слова

**Правила для имён переменных:**

- Допустимы **буквы** (латиница), **цифры** и символ **`_`**.
- Имя **не может начинаться с цифры**.
- Регистр важен: `age` и `Age` — **разные** имена.
- Нельзя использовать **ключевые слова** языка (см. список ниже): они зарезервированы под команды Python.

**Хороший стиль:** имя по смыслу (`total_score`, а не `ts`); составные имена пишутся как слова через подчёркивание (`user_age`, `first_name`).

**Примеры хороших имён:** `name`, `last_name`, `name_2`, `player_score`, `student_age`.

**Примеры, которые дадут ошибку или путаницу:** `2name` (цифра в начале), `last-name` (дефис), `my name` (пробел), имя из ключевых слов вроде `class` или `if`.
