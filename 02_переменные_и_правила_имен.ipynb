# 02. Переменные и имена

В этом уроке разберем, как хранить данные в переменных и как давать им понятные имена.

Это основа почти любого кода: сначала ты что-то сохраняешь в переменную, потом используешь это в вычислениях, условиях и играх.

## 1) Что такое переменная

Переменная - это "подписанная коробка" для данных.

Например:
- `name` хранит имя,
- `age` хранит возраст.

```python
name = "Вован"
age = 13
print(name)
print(age)
```

## 2) Пример с именем и фамилией

Здесь главное увидеть, что переменная может менять значение.

Сначала:
- `name = "Вован"`
- `last_name = "Вованов"`
- `name_2 = "Аня"`
- `last_name_2 = "Цветова"`

Потом меняем значение `last_name_2`.

```python
name = "Вован"
last_name = "Вованов"

name_2 = "Аня"
last_name_2 = "Цветова"

print("До:")
print(name, last_name)
print(name_2, last_name_2)

# Меняем значение переменной. Для девушек добавляем "a" к концу фамилии.
last_name_2 = last_name + 'a'

print("\nПосле:")
print(name_2, last_name_2)
```
