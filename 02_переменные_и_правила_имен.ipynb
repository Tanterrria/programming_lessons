## 1) Что такое переменная

Переменная - это "подписанная коробка" для хранения данных.

Пример:
- коробка с подписью `name` хранит имя,
- коробка с подписью `age` хранит возраст.

```python
name = "Вован"
age = 13
print(name)
print(age)
```

## 2) Твой пример про имена и фамилии

Идея: переменная может менять значение.

Исходно:
- `name = "Вован"`
- `last_name = "Вованов"`
- `name_2 = "Аня"`
- `last_name_2 = "Цветова"`

После события (по сюжету задачи) меняем фамилию `last_name_2`.

```python
name = "Вован"
last_name = "Вованов"

name_2 = "Аня"
last_name_2 = "Цветова"

print("До:")
print(name, last_name)
print(name_2, last_name_2)

# Меняем значение переменной
last_name_2 = "Вованова"

print("\nПосле:")
print(name_2, last_name_2)
```

## 3) Как можно называть переменные

### Хорошие варианты
- `name`
- `last_name`
- `name_2`
- `player_score`
- `student_age`

### Плохие варианты
- `2name` (нельзя начинать с цифры)
- `last-name` (дефис `-` запрещен)
- `my name` (пробелы запрещены)
- `class` (зарезервированное слово Python)

### Памятка
1. Используй английские буквы, цифры и `_`.
2. Не начинай с цифры.
3. Имя должно быть понятным по смыслу.
4. Лучше стиль `snake_case`.

```python
# Мини-тренировка: исправь названия на правильные
city_name = "Москва"
student_count = 25
is_raining = False
print(city_name, student_count, is_raining)
```
