# 03. Типы данных в Python

В Python у каждого значения есть **тип**. От типа зависит, какие **операции** можно делать и как данные хранятся в памяти.

**Динамическая типизация** — важная особенность Python: у одной и той же переменной тип может **меняться** по ходу программы. Это делает язык гибким, но нужно следить, что ты складываешь и сравниваешь.

В этом уроке — встроенные типы, операции с ними, приоритет вычислений, проверка и преобразование типов.

## Основные встроенные типы

Python поддерживает много встроенных типов. Каждый тип задаёт набор операций и способ хранения в памяти.

Узнать тип значения или переменной можно функцией **`type()`**:

```python
# Примеры разных типов данных
a = 42          # int — целое число
b = 3.14        # float — число с плавающей точкой
c = "Привет"    # str — строка
d = [1, 2, 3]   # list — список
e = (1, 2, 3)   # tuple — кортеж
f = {1, 2, 3}   # set — множество
g = {"ключ": "значение"}  # dict — словарь
h = True        # bool — логический тип

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
print(type(h))
```

### Арифметические операции с числами

| Оператор | Значение |
|----------|----------|
| `+` | сложение |
| `-` | вычитание |
| `*` | умножение |
| `/` | деление (результат **всегда** `float`) |
| `//` | целочисленное деление |
| `%` | остаток от деления |
| `**` | возведение в степень |

**Особенности деления:**
- `10 / 5` даёт `2.0`, а не `2`.
- `//` для отрицательных чисел округляет **вниз**.

```python
a = 15
b = 4

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)

print("10 / 5 =", 10 / 5)
print("10 // 3 =", 10 // 3)
print("-10 // 3 =", -10 // 3)
```

### Приоритет операций

Порядок как в математике (без скобок):

1. **Степень** `**` — для нескольких `**` подряд считают **справа налево**.
2. **Унарные** `+` и `-` (знак числа).
3. **Умножение и деление**: `*`, `/`, `//`, `%` — слева направо.
4. **Сложение и вычитание** — слева направо.

Скобки `()` меняют порядок.

```python
print(2 + 3 * 4)
print((2 + 3) * 4)
print(2 ** 3 ** 2)
print(-2 ** 2)
print((-2) ** 2)
print(10 / 3 + 2 * 4)
print(10 / (3 + 2) * 4)
```

### Операции со строками

- **`+`** — склеивание строк.
- **`*` на целое** — повторение.
- **`s[i]`**, **`s[start:end]`** — символ и срез.

Строку и число одним `+` не сложить без преобразования типа.

```python
first = "Привет"
second = "мир"
print(first + " " + second)

print("ха" * 3)
print("-" * 20)

word = "Python"
print(word[0], word[-1], word[2:5])
print("Py" in word, "Java" in word)
print(len(word))
```

### `isinstance()` — предпочтительнее `type()` для проверок

`isinstance()` удобна, когда допустимо несколько типов. В отличие от `type() == ...`, она учитывает **наследование** классов (пригодится в ООП).

```python
numbers = [1, 2, 3]
print(isinstance(numbers, list))
print(isinstance(numbers, (list, tuple)))
print(isinstance("Python", str))
print(isinstance(42, (int, float)))
```

### Пример обработки по типу

```python
def process_value(value):
    if isinstance(value, int):
        return f"Целое: {value * 2}"
    elif isinstance(value, float):
        return f"Дробное: {value:.2f}"
    elif isinstance(value, str):
        return f"Строка ({len(value)}): {value}"
    elif isinstance(value, list):
        return f"Список из {len(value)} элементов"
    return f"Другой тип: {type(value)}"

print(process_value(42))
print(process_value(3.14159))
print(process_value("Привет"))
print(process_value([1, 2, 3]))
```
