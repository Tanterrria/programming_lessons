# 03. Математика в Python

Здесь собраны базовые операции, которые нужны почти в любой программе: считать очки, здоровье, попытки, время и многое другое.

## Сложение `+`
Складывает числа.

```python
a = 7
b = 5
print("7 + 5 =", a + b)
```

## Вычитание `-`
Вычитает одно число из другого.

```python
a = 10
b = 3
print("10 - 3 =", a - b)
```

## Умножение `*`
Умножает числа.

```python
a = 4
b = 6
print("4 * 6 =", a * b)
```

## Степень `**`

`2 ** 3` читается как "два в степени три".
Это то же самое, что `2 * 2 * 2`.

```python
a = 2
b = 5
print("2 ** 5 =", a ** b)
```

## Пример из мини-игры

Представь, что это параметры персонажа. Посчитаем несколько значений.

```python
health = 10
bonus = 2
damage = 3

print("Здоровье + бонус:", health + bonus)
print("Здоровье - урон:", health - damage)
print("Урон * бонус:", damage * bonus)
print("Бонус в степени 3:", bonus ** 3)
```
