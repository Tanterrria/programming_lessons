# 03. Операции в Python и типы данных

В Python от **типа** значения зависит, какие **операции** допустимы. У целых и дробных чисел (`int`, `float`) — арифметика; у строк — другие операции; у списков и словарей — свои (разберём позже).

Ниже — базовые **числовые** операции для игр и расчётов: очки, здоровье, попытки, время. В конце — коротко про **строки** и **порядок** вычислений.

## Разные типы — разные операции

- **`int`** и **`float`**: сложение, вычитание, умножение, обычное деление, целочисленное деление, остаток от деления, степень.
- **`str` (строки)**: например склеивание через `+` и повторение через `*` на целое число — см. раздел про строки ниже.

Если смешать типы без преобразования (например сложить число и строку), Python выдаст ошибку — типы нужно согласовывать.

## Числа: сложение `+`

Складывает два числа (`int` или `float`).

```python
a = 7
b = 5
print("7 + 5 =", a + b)
```

## Числа: вычитание `-`

Вычитает одно число из другого.

```python
a = 10
b = 3
print("10 - 3 =", a - b)
```

## Числа: умножение `*`

Умножает два числа.

```python
a = 4
b = 6
print("4 * 6 =", a * b)
```

## Числа: деление `/`, целочисленное деление `//`, остаток `%`

- **`/`** — обычное деление; результат **всегда** с типом `float`, даже если делится нацело.
- **`//`** — целочисленное деление (дробная часть отбрасывается; для отрицательных чисел округление идёт «вниз»).
- **`%`** — остаток от деления.

```python
a = 15
b = 4
print("a / b:", a / b)
print("a // b:", a // b)
print("a % b:", a % b)
print("10 / 5:", 10 / 5)
```

## Числа: степень `**`

`2 ** 3` читается как «два в степени три».
Это то же самое, что `2 * 2 * 2`.

Для `**` порядок особый: в выражении вроде `2 ** 3 ** 2` сначала считают правую степень (подробнее — в разделе про приоритет).

```python
a = 2
b = 5
print("2 ** 5 =", a ** b)
```

## Приоритет операций

В одной строке без скобок порядок такой:

1. Степень `**` (для нескольких `**` подряд в Python считают **справа налево**).
2. Умножение и деление: `*`, `/`, `//`, `%`.
3. Сложение и вычитание: `+`, `-`.

Скобки `()` меняют порядок — как в математике.

```python
print(2 + 3 * 4)
print((2 + 3) * 4)
print(2 ** 3 ** 2)
print(-2 ** 2)
print((-2) ** 2)
```

## Строки (`str`): склеивание и повторение

У строк другие операции, чем у чисел:

- **`+`** между двумя строками **склеивает** их (конкатенация).
- **`*`** с целым числом **повторяет** строку столько раз.

`"a" * 3` даёт `"aaa"`. Смешивать строку и число одним `+` нельзя — сначала нужно преобразовать тип.

```python
greet = "Привет"
name = "Марс"
line = greet + ", " + name + "!"
print(line)
print("ха" * 3)
print("-" * 20)
```

## Пример из мини-игры

Представь, что это параметры персонажа. Посчитаем несколько значений.

```python
health = 10
bonus = 2
damage = 3

print("Здоровье + бонус:", health + bonus)
print("Здоровье - урон:", health - damage)
print("Урон * бонус:", damage * bonus)
print("Бонус в степени 3:", bonus ** 3)
print("Сколько раз урон «укладывается» в здоровье (//):", health // damage)
print("Остаток после деления здоровья на урон (%):", health % damage)
```
