# 03. Математические операции в Python

## Цель урока
Научиться выполнять основные математические действия в коде.

## После урока ученик умеет
1. Складывать, вычитать, умножать числа в Python.
2. Возводить число в степень.
3. Применять эти действия для простых задач.

## 1) Сложение `+`
Складывает числа.

```python
a = 7
b = 5
result = a + b
print("7 + 5 =", result)
```

## 2) Вычитание `-`
Вычитает одно число из другого.

```python
a = 10
b = 3
result = a - b
print("10 - 3 =", result)
```

## 3) Умножение `*`
Умножает числа.

```python
a = 4
b = 6
result = a * b
print("4 * 6 =", result)
```

## 4) Степень `**`
`a ** b` значит "a в степени b".

Пример: `2 ** 3 = 8`, потому что `2 * 2 * 2 = 8`.

```python
a = 2
b = 5
result = a ** b
print("2 ** 5 =", result)
```

## Общий пример из игрового контекста
Пусть у персонажа:
- здоровье `health = 10`
- бонус `bonus = 2`
- урон `damage = 3`.

```python
health = 10
bonus = 2
damage = 3

print("Здоровье + бонус:", health + bonus)
print("Здоровье - урон:", health - damage)
print("Урон * бонус:", damage * bonus)
print("Бонус в степени 3:", bonus ** 3)
```

## Частые ошибки
1. Путают `*` и `**`.
2. Пишут текст вместо числа там, где нужна математика.
3. Забывают сохранить результат в переменную при необходимости.
