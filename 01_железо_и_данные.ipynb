## 1) Что такое "железо"

Компьютер — это электронное устройство, которое получает данные, обрабатывает их по программе, хранит результат и выводит его пользователю. Это не обязательно только ноутбук или ПК.

Если устройство может выполнять программу (инструкции), обрабатывать данные и выдавать результат — это компьютерное устройство. **Планшет, телефон и даже обычный калькулятор - это тоже компьютер!**

Компьютер = ввод -> обработка -> хранение -> вывод.

**Железо** - это физические части компьютера.

- **CPU (по-русски говорят "ЦПУ", это процессор)** - "мозг", выполняет команды. _Сокращение от Central Processing Unit._
- **RAM (по-русски говорят "РЭМ", оперативная память)** - хранит данные, пока программа (или процесс) запущена. _Сокращение от Random Access Memory._
- **SSD/HDD (по-русски говорят "эсесдэшник/хэдедэшник" с ударением на последний слог, диск)** - внутренний накопитель, хранит файлы долго, даже когда компьютер выключен. Флэшка тоже считается накопителем, но внешним переносным. _Сокращение от Solid-State Drive (твердотельный накопитель), Hard Disk Drive (жесткий диск)._
- **GPU (по-русски говорят "ГПУ", видеокарта)** - обрабатывает графику: игры, видео, 3D. _Сокращение от Graphics Processing Unit._
- **Материнская плата** - это как основа-конструктор компьютера: на ней “держится” всё остальное. Она помогает всем компонентам работать вместе и обмениваться данными.
- **Устройства ввода** - клавиатура, мышь, микрофон.
- **Устройства вывода** - монитор, колонки, принтер.

Аналогия с кухней:
- **CPU — шеф-повар**: читает рецепт (код) и решает, что делать по шагам.
- **RAM** — рабочий стол: на нем лежит всё, что нужно прямо сейчас.
- **SSD/HDD** — шкаф и холодильник: там продукты и заготовки хранятся долго.
- **GPU** — кондитер/оформитель: отвечает за красивую подачу и сложное оформление (графику).
- **Материнская плата** — сама кухня с подключенными зонами: место, где всё установлено и связано.
- **ОС** — администратор кухни: распределяет задачи и следит, чтобы всё работало вместе.

## 2) Что такое операционная система (ОС)

**Операционная система** - это главная программа, которая управляет всем компьютером.

Примеры ОС:
- Windows
- macOS
- Linux
- Android и iOS (на телефонах)

Что делает ОС:
1. Запускает программы.
2. Управляет файлами и папками.
3. Распределяет память и время процессора.
4. Позволяет программам "общаться" с устройствами (экран, клавиатура, сеть).

## 3) Где хранятся данные

- Пока программа работает: данные обычно в **RAM**.
- Когда нужно сохранить надолго: записываем на **диск** в файл.

Пример:
- Ты ввел имя в программе -> оно в RAM.
- Сохранил текстовый файл с результатом -> он на диске.
Можно создать файл с расширением .txt (о расширениях есть дальше), в нем ввести текст. Пока мы не нажнем кнопку "Сохранить", изменения не сохранятся на диск, а останутся в оперативной памяти и при закрытии программы без сохранения удалятся и потеряются навсегда.

```python
# Демонстрация: данные в переменной во время работы программы
player_name = "Вован"
score = 10
print("Игрок:", player_name)
print("Очки:", score)
```

Если закрыть ноутбук, то данные в переменной player_name и score удалятся. В следующий раз, когда мы откроем ноутбук, нужно будет ввести имя игрока и очки заново.

```python
print("Игрок:", player_name)
```

## 5) Что можно делать с файлами
Файл — это способ хранить данные на диске надолго.
С файлами можно делать три базовых действия:
1. **Сохранять** — создать новый файл или записать в него данные.
2. **Редактировать** — изменить существующий файл (добавить, исправить, удалить часть текста).
3. **Удалять** — убрать файл с диска, если он больше не нужен.

Примеры из жизни:
- написал сочинение и нажал "Сохранить";
- исправил ошибку в тексте и снова сохранил;
- удалил старый файл, чтобы не мешался.

Важно:
- данные в переменной живут во время работы программы;
- данные в файле живут дольше, пока файл не удалят.
- в одной папке не может быть одинакового названия у двух файлов.

Удаленный файл иногда можно восстановить, но не всегда. Это зависит от типа диска (SSD/HDD) и способа записи данных на них.

## 6) Расширения файлов: как понять, что внутри

У файла обычно есть имя и расширение:
`имя.расширение`

Расширение — это подсказка, какой это тип файла и чем его открывать.

Примеры самых важных расширений:

- `.txt` — обычный текст (заметки, простые записи).
- `.md` — markdown-текст (красивое форматирование заголовков, списков, кода).
- `.py` — Python-скрипт (программа на Python).
- `.ipynb` — Jupyter Notebook (блокнот с текстом и кодом).
- `.json` — структурированные данные в формате "ключ: значение".
- `.csv` — таблица (данные по строкам и столбцам, часто для Excel).
- `.jpg`, `.png` — изображения.
- `.mp3`, `.wav` — аудио.
- `.mp4` — видео.
- `.pdf` — документ для чтения и печати.
- `.zip` — архив (упакованные файлы).

Важно помнить:
- Расширение помогает ОС выбрать программу для открытия файла.
- Переименование расширения не всегда "превращает" один тип файла в другой.
  Например, если переименовать `photo.jpg` в `photo.txt`,
  это не сделает фото текстовым файлом.

## 7) Архивы и сжатие

Иногда файлов много, и отправлять их по одному неудобно.  
Поэтому обычно делают архив.

Архив:
- собирает много файлов в один.
- часто уменьшает размер.

Чаще всего встретишь:
- `.zip` — самый обычный вариант.
- `.rar` и `.7z` — тоже архивы.

Что важно:
- текстовые файлы сжимаются хорошо.
- фото и видео обычно сжимаются слабо (они уже сжаты).
- перед работой архив обычно нужно распаковать.

Пример:
у тебя папка проекта (внутри нее 12 файлов и 3 папки).  
Делаешь `project.zip` и отправляешь один файл.
